```python
import numpy  as np
import matplotlib.pyplot as plt

T300_LeadingLog = np.loadtxt('Fixed T 0.3 p from 0.8 to 4 Leading Log photon emission calc.txt')
T500_LeadingLog = np.loadtxt('Fixed T 0.5 p from 0.8 to 4 Leading Log photon emission calc.txt')
T300_LeadingOrder = np.loadtxt('Fixed T 0.3 p from 0.8 to 4 Leading Order photon emission calc.txt')
T500_LeadingOrder = np.loadtxt('Fixed T 0.5 p from 0.8 to 4 Leading Order photon emission calc.txt')

T300_LLmomentum = T300_LeadingLog[:,0]
T300_LLrate = T300_LeadingLog[:,1]
T500_LLmomentum = T500_LeadingLog[:,0]
T500_LLrate = T500_LeadingLog[:,1]

T300_LOmomentum = T300_LeadingOrder[:,0]
T300_LOrate = T300_LeadingOrder[:,1]
T500_LOmomentum = T500_LeadingOrder[:,0]
T500_LOrate = T500_LeadingOrder[:,1]

plt.plot(T300_LLmomentum,T300_LLrate,marker='.',color='r',label='LL T=300 MeV')
plt.plot(T500_LLmomentum,T500_LLrate,marker='.',linestyle='dashed',color='r',label='LL T=500 MeV')
plt.plot(T300_LOmomentum,T300_LOrate,marker='.',color='b',label='LO T=300 MeV')
plt.plot(T500_LOmomentum,T500_LOrate,marker='.',linestyle='dashed',color='b',label='LO T=500 MeV')

plt.xlabel('$p_T$ (GeV)')
plt.ylabel('$ E d\Gamma_\gamma/dp^3$')
plt.title('Leading Log vs Leading Order Photon Emission Spectra')
plt.yscale('log')
plt.legend()
plt.show()
#Compares nicely with existing data from Kapusta and Aurenche
#The drop off with Leading Log is barely visible, but it can be seen
#The exponential decay of the Leading Order is also barely visible, but still seen
```
